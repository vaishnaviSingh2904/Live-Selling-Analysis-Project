#### #Detecting and Handling  Outliers

## Method used- **Isolation Forest**:
A tree-based model that isolates outliers by creating partitions. It isolates observations by randomly selecting a feature and then randomly selecting a split value between the maximum and minimum values of the selected feature. The logic is that outliers are few and different, so they are easier to isolate.

##### Why I used this and not the traditional z-score?
As per z-score method-result i was getting

Number of outlier rows in each column:

{'num_reactions': 235, 'num_comments': 130, 'num_shares': 159, 'num_likes': 222, 'num_loves': 144, 'num_wows': 43, 'num_hahas': 66, 'num_sads': 58, 'num_angrys': 76}
to be precise total 1133 outlier 

whereas with isolation forest i got 353 outlier

```python
from sklearn.ensemble import IsolationForest
numerical_cols = ['num_reactions', 'num_comments', 'num_shares', 'num_likes',
                  'num_loves', 'num_wows', 'num_hahas', 'num_sads', 'num_angrys']

# Fit the Isolation Forest model
iso_forest = IsolationForest(contamination=0.05, random_state=42)  # Adjust contamination as needed
df['anomaly'] = iso_forest.fit_predict(df[numerical_cols])

# Identify outliers
outliers = df[df['anomaly'] == -1]

# Calculate the percentage of outliers
percentage_outliers = (len(outliers) / len(df)) * 100

print(f"Number of outliers detected: {len(outliers)}")
print(f"Percentage of outliers: {percentage_outliers:.2f}%")
```

### About Anomaly collumn-
The anomaly column in the context of the Isolation Forest algorithm is a column added to your DataFrame that indicates whether each row is considered an outlier or not. The Isolation Forest model assigns a label to each data point:

 1 indicates an inlier (normal data point).

-1 indicates an outlier (anomalous data point).

```python
df_cleaned = df[df['anomaly'] != -1]

# Drop the anomaly column
df_cleaned = df_cleaned.drop(columns=['anomaly'])

print("Number of entries after removing outliers:", len(df_cleaned))
```

### Method used-
#### *Kernel Density Estimation (KDE) Plots*
KDE plots are used to estimate the probability density function of a continuous random variable. They provide a smooth curve that helps visualize the shape of the distribution.

##### How Kernel Density Estimation Works:

Kernel Function: KDE uses a kernel function (usually Gaussian) that assigns a weight (or influence) to each data point. This kernel function determines the shape and smoothness of the estimated density curve.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot KDE plots with fill=True
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.kdeplot(df_cleaned['num_reactions'], fill=True)
plt.title('Distribution of Number of Reactions (KDE)')
plt.xlabel('Number of Reactions')
plt.ylabel('Density')

plt.subplot(2, 2, 2)
sns.kdeplot(df_cleaned['num_comments'], fill=True)
plt.title('Distribution of Number of Comments (KDE)')
plt.xlabel('Number of Comments')
plt.ylabel('Density')

plt.subplot(2, 2, 3)
sns.kdeplot(df_cleaned['num_shares'], fill=True)
plt.title('Distribution of Number of Shares (KDE)')
plt.xlabel('Number of Shares')
plt.ylabel('Density')

plt.subplot(2, 2, 4)
sns.kdeplot(df_cleaned['num_likes'], fill=True)
plt.title('Distribution of Number of Likes (KDE)')
plt.xlabel('Number of Likes')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

### STEP 6:Clustering Analysis:

#### **ALOGORITHMS I USED**

#### K-Means
K-Means is a simple and efficient algorithm that partitions data into K clusters based on Euclidean distance.It's sensitive to initial placement of centroids and can be affected by outliers.Suitable for large datasets and can be used for both clustering and anomaly detection.

#### Hierarchical Clustering
Hierarchical Clustering builds a hierarchy of clusters by merging or splitting existing clusters based on a linkage criterion.It's useful for visualizing the data structure and can handle varying densities and shapes of clusters.Computationally expensive for large datasets, but provides interpretable results.

#### DBSCAN
DBSCAN is a density-based algorithm that groups data points into clusters based on their density and proximity.It's robust to noise and outliers, and can handle varying densities and shapes of clusters.Requires careful selection of epsilon (ε) and minimum samples (min_samples) parameters.



#### Expectation-Maximization (EM) Clustering
 EM Clustering is a probabilistic algorithm that models the data as a mixture of Gaussian distributions.It's useful for clustering and density estimation, and can handle high-dimensional data.Computationally expensive and requires careful selection of the number of components (K) and covariance type.



#### A)K-MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


# Define the numerical features
numerical_features = ['num_reactions', 'num_comments', 'num_shares', 
                      'num_likes', 'num_loves', 'num_wows', 
                      'num_hahas', 'num_sads', 'num_angrys', 'total_engagement']

# Select the relevant columns for clustering
X = df2[numerical_features]

# Initialize KMeans with the number of clusters (e.g., k=3)
kmeans = KMeans(n_clusters=3, n_init=25, random_state=1234, max_iter=300)

# Fit KMeans model to data
kmeans.fit(X)

# Add cluster labels to the original dataset
df2['cluster'] = kmeans.labels_

# Calculate and display cluster centers
cluster_centers = pd.DataFrame(kmeans.cluster_centers_, columns=X.columns)
print("Cluster Centers:")
print(cluster_centers)

# Set the figure size
plt.figure(figsize=(12, 8))

# Plot the Clusters
ax = sns.scatterplot(data=df2,
                     x='num_reactions',
                     y='num_comments',
                     hue='cluster',
                     palette='viridis',
                     alpha=0.8,
                     s=100)

# Plot the Centroids
ax = sns.scatterplot(data=cluster_centers,
                     x='num_reactions',
                     y='num_comments',
                     hue=cluster_centers.index,
                     palette='Set1',
                     s=500,
                     marker='D',
                     ec='black',
                     legend=False)

# Add Centroid Labels
for i in range(len(cluster_centers)):
    plt.text(x=cluster_centers.num_reactions[i],
             y=cluster_centers.num_comments[i],
             s=i,
             horizontalalignment='center',
             verticalalignment='center',
             size=15,
             weight='bold',
             color='white')

plt.title('K-Means Clustering')
plt.xlabel('num_reactions')
plt.ylabel('num_comments')
plt.legend(loc='upper right', title='Cluster')
plt.show()
```

**Cluster 0: High Reactions, Low Comments (Purple)**

### STEP 7: CLUSTER MODEL EVALUATION 

#### There are different parameters to evaluate models-

1)**Silhouette Score**: Measures how similar an object is to its own cluster compared to other clusters.

2)**Davies-Bouldin Index**: Measures the average similarity ratio of each cluster with its most similar cluster. Lower values indicate better clustering.

3)**Calinski-Harabasz Index**: Measures the ratio of the sum of between-clusters dispersion and of within-cluster dispersion. Higher values indicate better-defined clusters.
